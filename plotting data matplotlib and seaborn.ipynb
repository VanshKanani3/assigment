```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load Data:

```python
# Load the IRIS dataset
iris_df = pd.read_excel('IRIS.xlsx')
iris_df.head()
```

### A. Creating and customizing line charts using Matplotlib

```python
iris_df.plot(x="Id",y="SepalLengthCm")
plt.xlabel("ID")
plt.ylabel("Sepallength")
```

## B. Visualizing relationships between two or more variables using scatter plots

```python
x=np.array([12,12,13,14,123,11,123,144])
y=np.array([1,187,123,143,122,111,123,145])
plt.scatter(x,y)
plt.show
```

## C). Studying distributions of variables using histograms & bar charts to

```python
plt.hist(data=iris_df, x='PetalLengthCm')
plt.title("HISTOGRAM")
plt.show()
```

## D. Visualizing two-dimensional data using heatmaps


```python
corr_matrix = iris_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(data=corr_matrix,)
plt.title("Heatmap")
plt.show()
```

## E. Displaying images using Matplotlib's plt.imshow 


```python
random=np.random.random((10,10))
plt.imshow(random)
plt.title('Random Image')
plt.colorbar()
plt.show()
```
